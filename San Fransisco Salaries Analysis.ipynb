## Data Cleaning

Notes and Status Columns are empty in all rows so I will delete them

```python
df.drop(columns=['Notes', 'Status'], axis=1, inplace=True)
df.columns
```

Benefits and BasePay missing values will be imputed with the mean:

```python
# df['Benefits'].replace(np.nan, 0, inplace=True)
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)

```

OvertimePay and OtherPay have only 4 empty values so deleting their rows will not affects the data

```python
df.dropna(subset=['OvertimePay', 'OtherPay'], inplace=True)
```

## Grouped Analysis

### JobTitle

```python
# Convert relevant columns to string before applying the aggregation
df2= df.copy()
string_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']
df2[string_columns] = df2[string_columns].astype(str)
salary_groups2= df2.groupby(['JobTitle'] ,as_index=False).agg(
    {'JobTitle': 'first',

     'BasePay': ', '.join,
     'OvertimePay': ', '.join,
     'OtherPay': ', '.join,
     'Benefits': ', '.join,
     'TotalPay': ', '.join,
     'TotalPayBenefits': ', '.join,
     })
salary_groups2.head()
```

```python
mean_salary_groups_for_job_title= df.groupby(['JobTitle'],as_index=False)['TotalPay'].mean()
mean_salary_groups_for_job_title.head()
```

```python
mean_salary_groups_for_job_title.describe()
```

```python
sorted_mean_salary_groups = mean_salary_groups_for_job_title.sort_values(by='TotalPay', ascending=False)
sorted_mean_salary_groups.head(10)
```

## Simple Correlation Analysis

```python
def check_correlation(first_col,sec_col):
  correlation = df[first_col].corr(df[sec_col])

  # Plot a scatter plot to visualize the relationship
  plt.scatter(df[first_col], df[sec_col])
  plt.title(f'Scatter Plot: {first_col} vs {sec_col} (Correlation: {correlation:.2f})')
  plt.xlabel(first_col)
  plt.ylabel(sec_col)
  plt.show()
```

```python
check_correlation('TotalPay','TotalPayBenefits')
```

```python
check_correlation('TotalPay','BasePay')
```

```python
check_correlation('TotalPay','OvertimePay')
```

```python
check_correlation('TotalPay','OtherPay')
```

```python
check_correlation('TotalPay','Benefits')
```
